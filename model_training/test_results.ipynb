```python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_saver(saver):
    df = pd.read_csv(f'output/{saver}.csv')
    #display(df.head())
    best_f1 = df['Avg F1'].idxmax()
    print(df.iloc[best_f1])
    return df

df1 = visualize_saver('test1.1')
print('______')
df2 = visualize_saver('test1.2')
```

```python
def plot(df, metric, metric2, metric3):
    plt.figure(figsize=(10, 6))

    epochs = df['Epoch']
    feature = df[metric]
    feature2 = df[metric2]
    feature3 = df[metric3]

    plt.plot(epochs, feature, marker='o', label=metric)
    plt.plot(epochs, feature2, color='red', label=metric2)
    plt.plot(epochs, feature3, color='green', label=metric3) 

    # find max value and its epoch
    max_val = feature.max()
    max_epoch = epochs[feature.idxmax()]
    
    # annotate max value
    plt.annotate(f'{max_val:.2f}',
                    xy=(max_epoch, max_val),
                    xytext=(max_epoch, max_val + 0.02),
                    ha='center',
                    fontsize=9)
    
    
    # final plot 
    plt.title(f'{metric} across epoch')
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.ylim(0,1)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot(df2, 'Avg F1', 'Species F1', 'Colors F1')
```
